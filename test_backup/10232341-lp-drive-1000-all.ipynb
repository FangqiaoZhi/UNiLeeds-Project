```python

import matplotlib.pyplot as plt

fig, ax1 = plt.subplots()
plt.plot([3*p for p in range(5)] ,[1,3,5,7,9], 'r', label = 'training loss', )
#plt.plot([9,7,5,3,1], 'g', label = 'validation loss' )
plt.legend(loc='center')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.title('Training and validation loss and validation accuracy')
ax2=ax1.twinx()
ax2.plot([2,4,6,8,10], 'b', label = 'validation accuracy')
ax2.set_ylabel('accuracy')
plt.legend(loc='upper left')
plt.show()
```
